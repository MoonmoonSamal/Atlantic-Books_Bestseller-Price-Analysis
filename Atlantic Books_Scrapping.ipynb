# Web_Scrapping

## Atlantic Books - Non_Fiction

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
from tqdm import tqdm
```

```python
driver = webdriver.Chrome()
base_url = 'https://atlanticbooks.com/collections/non-fiction'
driver.get(base_url)
time.sleep(3)
```

```python
html_data = BeautifulSoup(driver.page_source, 'html.parser')
pages = int(html_data.find('a', {'class': 'pagination--item'}).text)
pages
```

```python
Titles = []
Authors = []
Prices = []
Original_Prices = []
Genres = []
Discounts = []
Descriptions = []
```

```python
for i in tqdm(range(pages)):
    url = f'https://atlanticbooks.com/collections/non-fiction?page={i + 1}'
    driver.get(url)
    time.sleep(3)
    html_data = BeautifulSoup(driver.page_source, 'html.parser')
    products = html_data.find_all('div', class_='productitem')
    
    for product in products:
        try:
            title = product.find('h2', {'class': 'productitem--title'}).text.strip() if product.find('h2', {'class': 'productitem--title'}) else 'No Title'
            Titles.append(title)
            
            author = product.find('p', {'class': 'clp-product-author'}).text.strip() if product.find('p', {'class': 'clp-product-author'}) else 'No Author'
            Authors.append(author)
            
            price = product.find('div', {'class': 'price__current price__current--on-sale'}).find('span', {'class': 'money'}).text.strip().replace('Rs.', '').strip() if product.find('div', {'class': 'price__current price__current--on-sale'}) else '0'
            Prices.append(price)
            
            original_price = product.find('div', {'class': 'price__compare-at--hidden'}).find('span', {'class': 'money'}).text.strip().replace('Rs.', '').strip() if product.find('div', {'class': 'price__compare-at--hidden'}) else '0'
            Original_Prices.append(original_price)

            discount = product.find('span', {'data-price-percent-saved': True}).text.strip()+"%" if product.find('span', {'data-price-percent-saved': True}) else '0'
            Discounts.append(discount)
            
            description = product.find('div', {'class': 'productitem--description'}).text.strip() if product.find('div', {'class': 'productitem--description'}) else 'No Description'
            Descriptions.append(description)
            
            Genres.append("Non-Fiction")
        
        except Exception as e:
            print(f"Error while extracting data from a product: {e}")

driver.quit()
```

```python
data = {
    'Title': Titles,
    'Price': Prices,
    'Original_Price': Original_Prices,
    'Discount': Discounts,
    'Author': Authors,
    'Genres': Genres,
    'Description': Descriptions
}

df = pd.DataFrame(data)
df
```

## Atlantic Books -Fiction

```python
driver1 = webdriver.Chrome()
base_url1 = 'https://atlanticbooks.com/collections/fiction'
driver1.get(base_url1)
time.sleep(3)
```

```python
html_data1 = BeautifulSoup(driver1.page_source, 'html.parser')
pages1 = int(html_data1.find('a', {'class': 'pagination--item'}).text)
pages1
```

```python
Title = []
Author = []
Price = []
Original_Price = []
Genre = []
Discount = []
Description = []
```

```python
for i in tqdm(range(pages1)):  
    url1 = f'https://atlanticbooks.com/collections/fiction?page={i + 1}' 
    driver1.get(url1)  
    time.sleep(3) 
    html_data1 = BeautifulSoup(driver1.page_source, 'html.parser')
    products = html_data1.find_all('div', class_='productitem') 
    
    for product in products:
        try:
            title = product.find('h2', {'class': 'productitem--title'}).text.strip() if product.find('h2', {'class': 'productitem--title'}) else 'No Title'
            Title.append(title)
            author = product.find('p', {'class': 'clp-product-author'}).text.strip() if product.find('p', {'class': 'clp-product-author'}) else 'No Author'
            Author.append(author)
            price = product.find('div', {'class': 'price__current price__current--on-sale'}).find('span', {'class': 'money'}).text.strip().replace('Rs.', '').strip() if product.find('div', {'class': 'price__current price__current--on-sale'}) else 'No Price'
            Price.append(price)
            original_price = product.find('div', {'class': 'price__compare-at--hidden'}).find('span', {'class': 'money'}).text.strip().replace('Rs.', '').strip() if product.find('div', {'class': 'price__compare-at--hidden'}) else '0'
            Original_Price.append(original_price)
            discount = product.find('span', {'data-price-percent-saved': True}).text.strip() + "%" if product.find('span', {'data-price-percent-saved': True}) else 'No Discount'
            Discount.append(discount)
            description = product.find('div', {'class': 'productitem--description'}).text.strip() if product.find('div', {'class': 'productitem--description'}) else 'No Description'
            Description.append(description)
            Genre.append("Fiction")
        
        except Exception as e:
            print(f"Error while extracting data from a product: {e}")

driver1.quit() 
```

```python
data1 = {
    'Title': Title,
    'Price': Price,
    'Original_Price': Original_Price,
    'Discount': Discount,
    'Author': Author,
    'Genres': Genre,
    'Description': Description
}

df1 = pd.DataFrame(data1)
df1
```
